```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("marvel_box_office.csv")
```

```python
data = df.fillna("No Phase")
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data.isnull(), cmap='viridis', cbar=False, yticklabels=False)
plt.title('Null Values Heatmap')
plt.show()
```

```python
data.plot()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot = True)
plt.plot()
```

```python
movies_per_year = data['Release Year'].value_counts().sort_index()

# Plot the histogram
plt.bar(movies_per_year.index, movies_per_year.values)

# Set the labels for x and y axes
plt.xlabel('Year')
plt.ylabel('Number of Movies')

# Set the title of the plot
plt.title('Number of Marvel Movies Released Each Year')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
movies_per_phase = data['Phase'].value_counts().sort_index()

# Plot the histogram
plt.bar(movies_per_phase.index, movies_per_phase.values)

# Set the labels for x and y axes
plt.xlabel('Phase')
plt.ylabel('Number of Movies')

# Set the title of the plot
plt.title('Number of Marvel Movies in Each Phase')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
movies_per_director = data['Director'].value_counts().sort_index()

plt.figure(figsize=(10, 8))

# Plot the histogram
plt.bar(movies_per_director.index, movies_per_director.values)

# Set the labels for x and y axes
plt.xlabel('Year')
plt.ylabel('Number of Movies')

# Set the title of the plot
plt.title('Number of Marvel Movies Released per Director')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Show the plot
plt.show()
```

```python
movies_per_month = data['Release Month'].value_counts().sort_index()

# Plot the histogram
plt.scatter(movies_per_month.index, movies_per_month.values)

# Set the labels for x and y axes
plt.xlabel('Year')
plt.ylabel('Number of Movies')

# Set the title of the plot
plt.title('Number of Marvel Movies Released Each Month')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
unique = data["Ownership"].value_counts()
unique
```

```python
labels = data['Ownership'].unique()

# Calculate the count of each unique value
ownership_counts = data['Ownership'].value_counts()

# Plot the pie chart
plt.figure(figsize=(8, 8))  # Adjust the figure size if needed
plt.pie(ownership_counts, labels=labels, autopct='%1.1f%%', startangle=140)

# Set the title of the plot
plt.title('Ownership of Movies')

# Show the plot
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
condition_high = df['Domestic Box Office'] > 500000000
condition_mid = (df['Domestic Box Office'] <= 500000000) & (df['Domestic Box Office'] > 100000000)

# Group the movies based on IMDb scores
high_imdb_movies = df[condition_high]
mid_imdb_movies = df[condition_mid]
low_imdb_movies = df[~(condition_high | condition_mid)]

# Plot the histogram
plt.hist([high_imdb_movies['Release Year'], mid_imdb_movies['Release Year'], low_imdb_movies['Release Year']],
         bins=20, color=['green', 'blue', 'red'], label=['Blockbuster', 'Average', 'Flop'])

# Set the labels for x and y axes
plt.xlabel('Year')
plt.ylabel('Number of Movies')

# Set the title of the plot
plt.title('Distribution of Marvel Movies by Domestic Box Office')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
condition_excellent = df['Run Time In Minutes'] > 140
condition_good = (df['Run Time In Minutes'] <= 140) & (df['Run Time In Minutes'] > 120)

# Group the movies based on Rotten Tomato Audience Score
excellent_movies = df[condition_excellent]
good_movies = df[condition_good]
average_movies = df[~(condition_excellent | condition_good)]

# Plot the histogram for Excellent movies
plt.bar('Excellent', len(excellent_movies), color='green', label='Long')

# Plot the histogram for Good movies
plt.bar('Good', len(good_movies), color='blue', label='Medium')

# Plot the histogram for Average movies
plt.bar('Average', len(average_movies), color='red', label='Short')

# Set the labels for x and y axes
plt.xlabel('Rotten Tomato Audience Score')
plt.ylabel('Number of Movies')

# Set the title of the plot
plt.title('Distribution of Marvel Movies by Run Time in Minutes')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
